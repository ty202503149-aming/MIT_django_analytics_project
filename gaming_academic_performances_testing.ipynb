```python
path = "/content/drive/MyDrive/AL_RAJAB_AMING_TESTING_FINAL_EXAM/Gaming_Academic_Performance.csv"

df = pd.read_csv(path)

df.head()
df.info()
```

```python
df_model = df.drop(columns=['student_id'])
```

```python
categorical_cols = ['gender', 'gaming_genre']
```

```python
df_encoded = pd.get_dummies(df_model, columns=categorical_cols, drop_first=True)
```

```python
df_encoded.head()
df_encoded.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_addiction = df_encoded['addiction_score']
```

```python
X_addiction = df_encoded.drop(columns=['addiction_score', 'grades'])

X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(
    X_addiction, y_addiction,
    test_size=0.2,
    random_state=42
)

reg_addiction = LinearRegression()
reg_addiction.fit(X_train_a, y_train_a)

y_pred_a = reg_addiction.predict(X_test_a)

mse_a = mean_squared_error(y_test_a, y_pred_a)
r2_a = r2_score(y_test_a, y_pred_a)

print("Addiction regression MSE:", mse_a)
print("Addiction regression R^2:", r2_a)
```

```python
stress_map = {'Low': 0, 'Medium': 1, 'High': 2}
df_encoded['stress_level_num'] = df_encoded['stress_level'].map(stress_map)
```

```python
y_stress = df_encoded['stress_level_num']
X_stress = df_encoded.drop(columns=['addiction_score', 'grades', 'stress_level', 'stress_level_num'], errors='ignore')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X_stress, y_stress, test_size=0.2, random_state=42, stratify=y_stress
)

clf_stress = LogisticRegression(max_iter=1000)
clf_stress.fit(X_train_s, y_train_s)

y_pred_s = clf_stress.predict(X_test_s)

print("Accuracy:", accuracy_score(y_test_s, y_pred_s))
print(classification_report(y_test_s, y_pred_s))
```

```python
stress_map = {'Low': 0, 'Medium': 1, 'High': 2}
df_encoded['stress_level_num'] = df_encoded['stress_level'].map(stress_map)
```

```python
y_stress = df_encoded['stress_level_num']
X_stress = df_encoded.drop(columns=['addiction_score', 'grades', 'stress_level', 'stress_level_num'], errors='ignore')
```

Next we will train:

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X_stress, y_stress, test_size=0.2, random_state=42, stratify=y_stress
)

clf_stress = LogisticRegression(max_iter=1000)
clf_stress.fit(X_train_s, y_train_s)

y_pred_s = clf_stress.predict(X_test_s)

print("Accuracy:", accuracy_score(y_test_s, y_pred_s))
print(classification_report(y_test_s, y_pred_s))
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

clf_stress = make_pipeline(
    StandardScaler(),
    LogisticRegression(max_iter=3000)
)

clf_stress.fit(X_train_s, y_train_s)
y_pred_s = clf_stress.predict(X_test_s)

print("Accuracy:", accuracy_score(y_test_s, y_pred_s))
print(classification_report(y_test_s, y_pred_s))
```

strong classification result overall. The model gets about 93.3% accuracy, which means it predicts the stress level correctly most of the time.
